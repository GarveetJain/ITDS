```python

# Part 2

# Train a machine learning model or pipeline that can take the raw strings from X and predict Y=0,1 depending on the
# sentiment of the tweet. Store the trained model in the variable `model`.
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
model = Pipeline([('countvectorizer', CountVectorizer()),('logisiticregression', LogisticRegression(max_iter=140))])
model.fit(X_train,Y_train)
```

```python

# Part 3

# Evaluate the model on the test set and calculate precision, and recall on both classes. Store the results in the
# variables `precision_0`, `precision_1`, `recall_0`, `recall_1`.
from sklearn.metrics import precision_score, recall_score
test_predictions = model.predict(X_test)
precision_0 = precision_score(Y_test,test_predictions, pos_label = 0)
precision_1 = precision_score(Y_test,test_predictions, pos_label = 1)
recall_0 = recall_score(Y_test,test_predictions, pos_label = 0)
recall_1 = recall_score(Y_test,test_predictions, pos_label = 1)
```
